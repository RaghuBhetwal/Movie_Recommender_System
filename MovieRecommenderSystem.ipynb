```python
#we have to convert string list to list

import ast
ast.literal_eval
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L 
```

```python
movies["genres"]= movies["genres"].apply(convert)
#for every movie we get the list of al the genres
```

```python
#Same for keywords
movies["keywords"]= movies["keywords"].apply(convert)
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert)
```

```python
#crew: we want director from dict
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L 
```

```python
movies["crew"] = movies["crew"].apply(fetch_director)
```

```python
#lets do it for genres
movies['genres']= movies['genres'].apply(lambda x : [i.replace(" ", "") for i in x])
```

```python
#Lets make a function and call it
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_movies_data = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
#Converting tags list to string

new_movies_data['tags'] = new_movies_data['tags'].apply(lambda x:" ".join(x))
new_movies_data.head()
```

```python
#Converting to lowercase

new_movies_data['tags'] = new_movies_data['tags'].apply(lambda x: x.lower())
```

```python
#Stemming
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer
```

```python
import nltk
nltk.download('punkt')
from nltk.stem import PorterStemmer

ps = PorterStemmer()

def stem(text):
    words = nltk.word_tokenize(text)
    stemmed_words = [ps.stem(word) for word in words]
    return " ".join(stemmed_words)
```

```python
new_movies_data['tags'] = new_movies_data['tags'].apply(stem)
```

```python
new_movies_data['tags']
```

```python
#Text Vectorzation
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')


#we take 5000 words
```

```python
vector = cv.fit_transform(new_movies_data['tags']).toarray()
```

```python
vector.shape
```

```python
vector
```

```python
cv.get_feature_names()
```

```python
#Finding the distance between each and every movie or words using Cosine Similarity
#we calclate the angle:: 0 degree is closet or in same line, 90 is high distance, 180 degree is complete is opposite
#distance is inverse proportion to Cosine Similarity

from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
similarity[0]

#the first movie has the 4806 cosine similarity with everyone
```

```python
similarity.shape
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key = lambda x: x[1]) [1:6]

#the first movie with top 5 cosine similarity distance
```

```python
def recommend(movie):       
    index = new_movies_data[new_movies_data['title'] == movie].index[0]
    movies_list = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])[1:6]
    for i in movies_list:
        print(new_movies_data.iloc[i[0]].title)
        
```

```python
recommend('Avatar')
```
